This tutorial is written for physics majors with little or no coding background. It starts from symbolic mathematics and numerical methods, then follows the scalar wave equation into generated C projects and introductory numerical relativity.

The notebooks are intended to be read with stored outputs visible on GitHub and rerun locally when desired. A **notebook cell** is one runnable or readable block in a notebook, and a **kernel** is the Python process that runs code cells.

Key software terms used throughout: a **package** is installable Python code; a **module** is one importable file or namespace inside a package; `import` loads that code into the kernel; a **function** is reusable code that accepts **arguments** and may produce a **return value**; an **object** is a Python value with data and behavior; and a **dictionary** stores key-value pairs.

Key numerical-code terms: a **parameter** is a named knob in a calculation; a **symbolic expression** is algebra represented by SymPy; a **grid** is a set of sample points; a **gridfunction** is a field sampled on those points; a **stencil** estimates a derivative from neighboring samples; a **right-hand side** gives the time derivative in an evolution equation; **generated code** is C text emitted from symbolic expressions; a **C function prototype** declares a C function to other files; and a **registry** is a named collection NRPy later walks over to write files.

Key build and infrastructure terms: a **compiler** translates C into machine code; `make` runs build rules; an **executable** is the compiled program; a **backend** is a code-writing target; **infrastructure** is support code around the physics kernels; **BHaH** is NRPy's standalone generated C backend; the **Einstein Toolkit** is a Cactus-based simulation framework; **ETLegacy** writes Einstein Toolkit-style files; a **thorn** is an Einstein Toolkit component; and **CCL** files declare thorn interfaces, parameters, schedules, and source lists.

Key relativity terms: **ADM** splits spacetime into fields on spatial slices; **BSSN** rewrites ADM variables for stable numerical evolutions; the **lapse** controls normal-time advance between slices; the **shift** controls coordinate motion within a slice; and **extrinsic curvature** measures how the slice bends in spacetime.
