# The NRPy 2-torial
### Lead author: [Zachariah B. Etienne](https://etienneresearch.com)

This tutorial is written for physics majors who know the mathematics before they know the software. It starts with symbolic expressions and finite differences, then follows the scalar wave equation into generated C projects and introductory numerical relativity.

If you are new to notebooks, a **notebook cell** is one readable or runnable block, and a **kernel** is the Python process that runs code cells. A **package** is installable Python code; a **module** is an importable part of a package; `import` loads it; a **function** accepts **arguments**; and an **environment** is the Python installation used by both the terminal and notebook kernel.

## Running the Notebooks

Start with the setup notebook, then open the notebooks in order. Later generated-project lessons need `make` and a C compiler; those tools are introduced only when a project is actually generated.


## Learning Goals

- Use the tutorial route as a sequence, not as a reference manual.
- Learn each idea by predicting what a code cell should do, running it, and explaining the result in words.
- Keep the physics question in view even when the notebook is showing code or generated files.

## Words for This Notebook

- **Notebook:** a document made of readable text cells and runnable Python cells.
- **Kernel:** the Python process that remembers variables while a notebook is open.
- **Generated code:** C code written by NRPy from symbolic formulas, so the computer can run the formulas quickly.
- **Diagnostic:** a number printed by a run so you can judge whether the calculation behaved as expected.

Use the code cells actively: first predict what should happen, then run the cell, then explain the output in plain language. This predict-run-explain pattern keeps the physics idea connected to the programming details.
