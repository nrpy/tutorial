***If you are unfamiliar with using Jupyter Notebooks, first review the official [Jupyter Notebook Basics Guide](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/Notebook%20Basics.html).***

This tutorial is written for physics majors with little or no coding background. It starts from symbolic mathematics and numerical methods, then follows the scalar wave equation into generated C projects and introductory numerical relativity.

Notebooks are committed with stored outputs so GitHub can display the calculations. You can rerun them locally when your Python environment has NRPy, Jupyter, `make`, and a C compiler.

Survival glossary: a **notebook cell** is one runnable or readable block, and a **kernel** is the Python process that runs code cells. A **package** is installable Python code; a **module** is an importable part of a package; `import` loads it; a **function** accepts **arguments** and may return a value; an **object** is a Python value with data and behavior; and a **dictionary** stores key-value pairs. A **parameter** is a named knob; a **symbolic expression** is algebra represented by SymPy; a **grid** is a set of sample points; a **gridfunction** is a sampled field; a **stencil** estimates derivatives from neighboring samples; a **right-hand side** gives a time derivative; **generated code** is C text emitted from symbolic expressions; a **C function prototype** declares a function; and a **registry** is a named collection NRPy later writes to files. A **compiler** turns C into machine code; `make` runs build rules; an **executable** is the compiled program; a **backend** is a generated-code target; **infrastructure** is support code around physics kernels; **BHaH** is NRPy's standalone generated C backend; the **Einstein Toolkit** is a Cactus-based simulation framework; **ETLegacy** writes Einstein Toolkit-style files; a **thorn** is an Einstein Toolkit component; and **CCL** files declare thorn interfaces, parameters, schedules, and source lists. **ADM** splits spacetime into fields on spatial slices; **BSSN** rewrites ADM variables for stable numerical evolution; the **lapse** controls normal-time advance; the **shift** controls coordinate motion within a slice; and **extrinsic curvature** measures how the slice bends in spacetime.
