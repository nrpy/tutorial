## Step 1: Verify the Imported Package

The path printed here is runtime context only. It tells you which NRPy package this kernel imported.

```
import importlib
import importlib.util
import inspect
import pkgutil

import nrpy

print(nrpy.__file__)
```

## Step 2: Discover Major Package Namespaces

The main NRPy namespaces separate runnable examples, symbolic equations, infrastructure backends, shared helpers, and expression-validation tools.

```
def package_children(package_name):
    module = importlib.import_module(package_name)
    if not hasattr(module, "__path__"):
        return []
    return sorted(
        item.name
        for item in pkgutil.iter_modules(module.__path__)
        if not item.name.startswith("_")
    )


package_areas = [
    "nrpy.examples",
    "nrpy.equations",
    "nrpy.infrastructures",
    "nrpy.helpers",
    "nrpy.validate_expressions",
]

for package_name in package_areas:
    children = package_children(package_name)
    print(package_name)
    for name in children:
        print(f"  {name}")
```

## Step 3: List Example Modules

`nrpy.examples` contains runnable project generators. Listing modules with `pkgutil` inspects the package directory without importing each generator.

```
example_modules = package_children("nrpy.examples")
for name in example_modules:
    print(name)
```

## Step 4: Check Representative Modules

Use `find_spec()` to check that selected current modules are discoverable before importing them.

```
representative_modules = {
    "Cartesian wave-equation generator": "nrpy.examples.wave_equation_cartesian",
    "wave-equation RHSs": "nrpy.equations.wave_equation.WaveEquation_RHSs",
    "BHaH infrastructure": "nrpy.infrastructures.BHaH",
}

for label, module_name in representative_modules.items():
    found = importlib.util.find_spec(module_name) is not None
    print(f"{label}: {module_name} -> {found}")
```

## Step 5: Inspect Wave-Equation RHS Expressions

Equation modules are safe to import directly when you want symbolic expressions. Here the wave-equation RHS object exposes the evolution equations for `uu` and `vv`.

```
from nrpy.equations.wave_equation.WaveEquation_RHSs import WaveEquation_RHSs

rhs = WaveEquation_RHSs()
print(inspect.getdoc(WaveEquation_RHSs))
print(f"uu_rhs = {rhs.uu_rhs}")
print(f"vv_rhs = {rhs.vv_rhs}")
```

## Step 6: Apply a Helper Function

Helper modules provide small utilities used across tutorials and package internals. This helper extracts the derivative symbols appearing in the RHS expression.

```
from nrpy.helpers.expression_utils import get_unique_expression_symbols_as_strings

symbols = get_unique_expression_symbols_as_strings(
    rhs.vv_rhs,
    exclude=["wavespeed"],
)
print(symbols)
```
