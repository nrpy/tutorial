## Import the Core Modules
The loop imports modules that are safe to load without generating files. Generator modules are run later with `python -m ...` so their side effects happen in disposable workspaces.

```python
import importlib

module_names = [
    "nrpy.params",
    "nrpy.indexedexp",
    "nrpy.grid",
    "nrpy.finite_difference",
    "nrpy.c_codegen",
    "nrpy.c_function",
    "nrpy.reference_metric",
    "nrpy.infrastructures.BHaH",
    "nrpy.infrastructures.ETLegacy",
]

for module_name in module_names:
    module = importlib.import_module(module_name)
    print(f"{module_name} -> {module.__name__}")
```

## Use the Modules Once
These direct calls show the style of later notebooks: parameters become symbolic names, indexed expressions organize components, and C generation emits assignments.

```python
import sympy as sp

import nrpy.c_codegen as ccg
import nrpy.indexedexp as ixp
import nrpy.params as par

wave_speed = par.register_CodeParameter("REAL", "tutorial_package_map", "package_map_wave_speed", 1.0)
vector = ixp.declarerank1("vU")
expr = wave_speed * (vector[0] + sp.Rational(1, 2) * vector[1])

print("parameter symbol:", wave_speed)
print("vector components:", vector)
print(ccg.c_codegen(expr, "rhs", include_braces=False, verbose=False))
```
